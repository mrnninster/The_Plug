```
# Testing Dependencies
import os
import cv2
import time 
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
%matplotlib inline
```

```
start = time.time()
print("hello")
end = time.time()
print(end - start)
```

```
# Function to resize input image to required size, Take image_path as input
def Image_prepare(ImagePath):
    IMG_SIZE = 128
    img_array = cv2.imread(ImagePath)
    img_array = img_array
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-3, IMG_SIZE, IMG_SIZE, 3)
```

```
# Test with TF-Keras model
model_path = "drive/My Drive/Plug/Tutorial_Dataset/hands2.1.h5"
Image_Directory = "drive/My Drive/Plug/Tutorial_Dataset/Held_Images"
Image_list = os.listdir("drive/My Drive/Plug/Tutorial_Dataset/Held_Images")

# Loading Keras Model
Loaded_Model = load_model(model_path)

# Classification Classes
Classes = ['3L', '1R', '0R', '2L', '1L', '2R', '3R', '4L', '0L', '4R', '5R', '5L']

# Loop Over all Images and Make Prediction
for image in Image_list[:3]:
  start = time.time()
  image_path = Image_Directory+"/"+image
  print(image_path)

  # Model Prediction
  Res = Loaded_Model.predict(Image_prepare(image_path))

  # Fetching Max Value and its Index Position From Prediction List
  Pred = max(Res[0])
  print(Pred)
  Result_List = Res[0].tolist()  #Convert to Python List
  Pos = Result_List.index(Pred)

  # Predicted Result
  print(Classes[Pos])
  end = time.time()
  print(end - start)
  print('')
```

```
# Test with TFLite Model
# This is provided in the Tensorflow Lite documentation
# please refer to the documentation if you want to find
# you find out more.

# Load the TFLite model and allocate tensors.
TFLite_Model = "drive/My Drive/Plug/Tutorial_Dataset/hands2_1.tflite"
interpreter = tf.lite.Interpreter(model_path=TFLite_Model)
interpreter.allocate_tensors()

# Get input and output tensors.
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Test the model on random input data.
input_shape = input_details[0]['shape']

# Loop Over all Images and Make Prediction
for image in Image_list[:3]:
  start = time.time()
  image_path = Image_Directory+"/"+image        #Image Path
  print(image_path)

  input_data = np.array(Image_prepare(image_path), dtype=np.float32)    #Creating Image Array
  interpreter.set_tensor(input_details[0]['index'], input_data)

  interpreter.invoke()

  # Infrence Result
  output_data = interpreter.get_tensor(output_details[0]['index'])

  # Fetching Max Value and its Index Position From Prediction List
  Pred = max(output_data[0])
  print(Pred)
  Result_List = output_data[0].tolist()  #Convert to Python List
  Pos = Result_List.index(Pred)

  # Predicted Result
  print(Classes[Pos])
  end = time.time()
  print(end - start)
  print('')
```
