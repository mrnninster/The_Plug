# CREATING CLASSES FOLDER

```python
# creating image classes for the train data
for i in range(0,6):
    os.mkdir('train/'+str(i)+('R'))
    os.mkdir('train/'+str(i)+('L'))
print('done')
```

```python
# creating image classes for the test data
for i in range(0,6):
    os.mkdir('test/'+str(i)+('R'))
    os.mkdir('test/'+str(i)+('L'))
print('done')
```

```python
#creating image classes for the validation data
for i in range(0,6):
    os.mkdir('validation/'+str(i)+('R'))
    os.mkdir('validation/'+str(i)+('L'))
print('done')
```

# SORTING IMAGES

### Training Images

```python
# Selecting Image and Moving Image for training images
files = os.listdir('train')
```

```python
for i in files:
    print(i)
```

```python
len(files)
```

```python
for image in range (0,14479):
    imagename = files[image]
    if len(imagename) > 5:
        split_name = imagename.split('.')
        print(split_name[0][-2:]) #Print Image Class
        file_class = split_name[0][-2:]
        os.replace('train/'+imagename,'train/'+file_class+'/'+imagename)
        print('moved '+str(imagename)+' to '+str(file_class))
        
    else:
        pass
```
